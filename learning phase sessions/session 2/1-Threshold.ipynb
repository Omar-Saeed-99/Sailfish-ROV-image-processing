```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

image.png

```python
img = cv2.imread('download5.png')
```

```python
cv2_imshow(img)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(gray)
```

```python
ret, thresh1 = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY)
```

```python
cv2_imshow(thresh1)
```

```python
print(thresh1)
```

```python
cv2_imshow(thresh2)
```
