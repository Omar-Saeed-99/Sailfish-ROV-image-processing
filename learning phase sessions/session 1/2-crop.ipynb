```python
plt.imshow(img)
```

We want to crop that image, and as we explained in the previous noteook, the image is an array, and when we want to crop an image, we choose the pixel range in which we want it to appear, just like indexing the array.

Example if we have a list [0,1,2,3,4,5]
And we want to specify the numbers from 1 to 3

```python
list = [0,1,2,3,4,5]
list[1:4]
```

This is exactly what we do in the Image

```python
img = img[200:500,100:400]
```

```python
plt.imshow(img)
```
