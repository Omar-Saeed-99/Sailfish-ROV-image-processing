```python
import numpy as np
import cv2 
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('thumbs_up_down.jpg')
```

```python
cv2_imshow(img)
```

```python
gray =  cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(gray)
```

```python
ret,binary = cv2.threshold(gray,220,255,cv2.THRESH_BINARY_INV)
```

```python
cv2_imshow(binary)
```

```python
contour , h = cv2.findContours(binary,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(img,contour,-1,(255,0,0),3);
```

```python
cv2_imshow(img)
```

```python
# 46 , 0 , 9 , 244 , 255 , 255
low = np.array([46 , 0 , 9])
high= np.array([224,255,255])
```

```python
mask = cv2.inRange(img,low,high)
```

```python
cv2_imshow(mask)
```

```python
contour , h = cv2.findContours(mask,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(img,contour,-1,(255,0,0),3);
```

```python
cv2_imshow(img)
```

```python
for cnt in contour:
  if cv2.contourArea(cnt)>1000:
    x,y,w,h = cv2.boundingRect(cnt)
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),5)
```

```python
cv2_imshow(img)
```
