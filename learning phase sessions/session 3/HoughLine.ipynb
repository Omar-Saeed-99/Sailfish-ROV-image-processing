```python
import cv2 
import matplotlib.pylab as plt 
import numpy as np 
```

```python
img = cv2.imread('2.png')
img = cv2.GaussianBlur(img,(5,5),cv2.BORDER_DEFAULT)
```

```python
plt.imshow(img)
```

```python
img = img[400:1000,400:1600]
```

```python
plt.imshow(img)
```

```python
edge = cv2.Canny(img,102, 109)
plt.imshow(edge)
```

```python
lines = cv2.HoughLinesP(edge, 2, np.pi/180, 153,minLineLength=100,maxLineGap=100 )
black = black2 =np.zeros(img.shape[0:2]).astype(np.uint8)
```

```python
for line in lines:
  x1,y1,x2,y2 = line[0]
  cv2.line(img,(x1,y1),(x2,y2),[0,0,255],1) 
  cv2.line(black,(x1,y1),(x2,y2),[255,0,255],1) 
```
